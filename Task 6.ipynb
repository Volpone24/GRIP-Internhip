```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from dtreeviz.trees import dtreeviz
```

```python
df = pd.read_csv("Iris.csv")
```

```python
df.head()
```

```python
X = df.iloc[:,[1,2,3,4]]
X.head()
```

```python
Y = df[['Species']]
print(Y)
```

```python
df.Species.unique()
```

```python
X.columns
```

```python
species_to_no = {'Species': {"Iris-setosa":0, "Iris-versicolor":1, "Iris-virginica":2}}
Y = Y.replace(species_to_no)
Y
```

```python
dtc = DecisionTreeClassifier(random_state=0)
dtc.fit(X,Y)
```

```python
plt.figure(figsize=(16,10))
featurename = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
classname = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
tree.plot_tree(dtc, feature_names = featurename, class_names=classname, filled=True)
plt.show()
```

```python
plt.figure(figsize=(16,7))
dtreeviz(dtc,X,Yfeature_names = featurename, class_names=classname)
plt.show()
```
